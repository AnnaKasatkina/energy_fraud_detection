```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('data_prepared.xlsx')
data
```

```python
m1_cols = [col for col in data.columns if col.startswith('m1_')]
data_m1 = data[m1_cols].copy()

# index = 0
# for cols in data_m1.columns:
#     new_name = 't' + str(index)
#     data_m1 = data_m1.rename(columns={cols: new_name})
#     index += 1

data_m1.tail(10)
```

```python
counter_index = 3
counter_data = data_m1.loc[counter_index]
time_points = range(len(counter_data))

plt.figure(figsize=(12,6))
plt.plot(time_points, counter_data.values, linewidth=.5)
```

```python
total_nans = int(data_m1.isna().sum().sum())
n_rows_with_nan = int((data_m1.isna().any(axis=1)).sum())
total_nans, n_rows_with_nan
```

```python
data_m1 = data_m1[data_m1.isna().mean(axis=1) <= 0.5]
data_m1.isna().sum().sum()
```

```python
data_m1 = data_m1.interpolate(axis=1, method='linear', limit_direction='both')
data_m1.reset_index(inplace=True)
data_m1.isna().sum().sum()
```

```python
data_m1.drop('index', axis=1, inplace=True)
data_m1.tail(10)
```

```python
for i in range(35):
    count_zero = (data_m1.iloc[i] == 0).sum()
    if count_zero > 0:
        print(i, count_zero)
```

```python
data_m1 = data_m1[data_m1.sum(axis=1) != 0].reset_index()
data_m1.drop('index', axis=1, inplace=True)
data_m1.tail()
```

```python
data_m1.to_csv('data.csv')
```
